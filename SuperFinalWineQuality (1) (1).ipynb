```python
#The head() method returns the first 5 rows if a number is not specified.
data.head(10)
```

```python
#The tail() method returns the last 5 rows if a number is not specified.
data.tail(10)
```

```python
#The info() method contains the number of columns, column labels, column data types, memory usage, range index, and the number of cells in each column(non-null values).
data.info()
```

```python
#The describe() method returns description of the data.
data.describe()
```

```python
#The shape function in python is used to find the dimensions of data structures.
data.shape
```

```python
#The isnull() function is used to find number of null values if present in data.
data.isnull().sum().sum()
```

```python
#The plot() function is used to draw points in a diagram.
data.plot()
```

```python
#The corr() method finds the correlation of each column in a data.
data.corr()
```

```python
#The heatmap is a great way to visualize data, because it can show the relation between variables including time.
sns.heatmap(data.corr(),center = True,cmap='hot')
```

```python
#The hist() method on a data will return histograms for all non-nuisance series in the data.
#bins is essentially the number of classes the histogram divides the data into. More bins will make the histogram smoother.
#figsize: The layout in which the output graphs must be, for example, (3,1) gives the figure in a single column and three rows.
data.hist(bins=30,figsize=(30,30))
```
